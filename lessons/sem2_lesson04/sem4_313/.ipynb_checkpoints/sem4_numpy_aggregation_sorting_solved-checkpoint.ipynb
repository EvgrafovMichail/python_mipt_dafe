```python
import json
import os

import matplotlib.pyplot as plt
import numpy as np
```

```python
class LogScaler:
    _name: str

    def __init__(self) -> None:
        self._name = "log-scaler"

    def __call__(self, array: np.ndarray) -> np.ndarray:
        array_scaled = array.copy()

        nonzeros = array != 0
        array_scaled[nonzeros] = np.log10(array[nonzeros])
        array_scaled[nonzeros] -= array_scaled[nonzeros].min()

        array_range = array_scaled.max() - array_scaled.min()
        array_scaled = (array_scaled - array_scaled.min()) / array_range

        return (255 * array_scaled).astype(np.uint8)
    
    @property
    def name(self) -> str:
        return self._name
```

```python
class RangeScaler:
    _name: str

    def __init__(self) -> None:
        self._name = "range-scaler"

    def __call__(self, array: np.ndarray) -> np.ndarray:
        array_scaled = array.copy()

        array_range = array_scaled.max() - array_scaled.min()
        array_scaled = (array_scaled - array_scaled.min()) / array_range

        return (255 * array_scaled).astype(np.uint8)
    
    @property
    def name(self) -> str:
        return self._name
```

```python
class RowRangeScaler:
    _name: str

    def __init__(self) -> None:
        self._name = "row-range-scaler"

    def __call__(self, array: np.ndarray) -> np.ndarray:
        array_scaled = array.copy()

        mins = array_scaled.min(axis=-1, keepdims=True)
        maxs = array_scaled.max(axis=-1, keepdims=True)
        ranges = maxs - mins

        mask = array_scaled.sum(axis=-1) != 0
        array_scaled[mask] = (array_scaled - mins)[mask] / ranges[mask]
        
        return (255 * array_scaled).astype(np.uint8)

    @property
    def name(self) -> str:
        return self._name
```

```python
path_to_radar_data = os.path.join("data", "radar_data.json")
assert os.path.exists(path_to_radar_data)

with open(path_to_radar_data, "r") as file:
    radar_data = np.array(json.load(file), dtype=np.float64)
```

```python
scalers = [RangeScaler(), RowRangeScaler(), LogScaler()]
images_scaled = {
    scaler.name: scaler(radar_data) for scaler in scalers
}
```

```python
_, axes = plt.subplots(len(images_scaled), 1, figsize=(16, 9))
pairs = list(images_scaled.items())
axes = axes.flatten()

for i, pair in enumerate(pairs):
    label, image = pair

    axes[i].set_title(label, fontsize=15, fontweight="bold", c="dimgray")
    axes[i].imshow(image.T, cmap="gray")
    axes[i].axis("off")
```

```python
import sklearn.datasets as skd
import matplotlib.pyplot as plt
import scipy.stats as sps
import numpy as np
```

```python
x, y = skd.make_moons(n_samples=200, noise=0.5)
reds = x[y == 0]
blues = x[y == 1]
```

```python
plt.figure(figsize=(16, 9))
plt.scatter(reds[..., 0], reds[..., 1], c="r")
plt.scatter(blues[..., 0], blues[..., 1], c="b")
```

```python
from typing import Union


class KNN:
    _k_neighbours: int
    _points: Union[np.ndarray, None]
    _labels: Union[np.ndarray, None]
    
    def __init__(self, k_neighbours: int = 5) -> None:
        self._k_neighbours = int(k_neighbours)
        self._points = None
        self._labels = None
        
    def fit(self, points: np.ndarray, labels: np.ndarray) -> None:
        self._points = points.copy()
        self._labels = labels.copy()
        
    def predict(self, points: np.ndarray) -> np.ndarray:
        if self._points is None or self._labels is None:
            raise RuntimeError
            
        differences = points[:, np.newaxis, :] - self._points[np.newaxis, ...]
        distances = np.sum(differences ** 2, axis=-1)
        
        indices = np.argsort(distances, axis=-1)[..., :self._k_neighbours]
        nearest_labels = self._labels[indices]
        modes, _ = sps.mode(nearest_labels, axis=-1)
        
        return np.array(modes).flatten()
```

```python
knn = KNN()
knn.fit(x, y)
knn.predict(x)
```

```python
_, axis = plt.subplots(figsize=(16, 9))

labels_unique = np.unique(y)

for label in labels_unique:
    abscissa = x[label == y]
    axis.scatter(abscissa[..., 0], abscissa[..., 1], label=label)
    
axis.legend()
```
