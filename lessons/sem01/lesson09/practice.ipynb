## Неправильное сравнение типов

**Пример**:

```python
string = "this is string"

if type(string) == str:
    print("var has str type")
```

**Исправление**:

```python
string = "this is string"

if isinstance(string, str):
    print("var has str type")
```

## Слишком большой уровень отступа

**Пример**:

```python
nums = list(range(10))

for num in nums:
    if num % 2 == 0:
        print(f"{num} dividable by 2")

        if num % 3 == 0:
            print(f"{num} dividable by 6")

            if num % 5 == 0:
                print(f"{num} dividable by 30")
```

**Исправление**:

```python
nums = list(range(10))

for num in nums:
    if num % 2 != 0:
        continue

    print(f"{num} dividable by 2")

    if num % 3 != 0:
        continue

    print(f"{num} dividable by 6")

    if num % 5 != 0:
        continue

    print(f"{num} dividable by 30")
```

## Дублирование логики

**Пример**:

```python
nums = list(range(10))

for num in nums:
    if num % 2 != 0:
        continue

    print(f"{num} dividable by 2")

    if num % 3 != 0:
        continue

    print(f"{num} dividable by 6")

    if num % 5 != 0:
        continue

    print(f"{num} dividable by 30")
```

**Исправление**:

```python
nums = list(range(10))
dividers: tuple[int, ...] = (2, 3, 5)

for num in nums:
    dividers_product = 1

    for divider in dividers:
        dividers_product *= divider

        if num % divider != 0:
            break

        print(f"{num} dividable by {dividers_product}")
```

## Магические константы

**Пример**:

```python
status_code = 200

if 400 <= status_code:
    print(f"HTTP error with code: {status_code}")

else:
    print("OK")
```

**Исправление**:

```python
from typing import Final


HTTP_CODE_BAD_REQUEST: Final[int] = 400

status_code = 200

if HTTP_CODE_BAD_REQUEST <= status_code:
    print(f"HTTP error with code: {status_code}")

else:
    print("OK")
```

## Неправильное ветвление в функциях

**Пример**:

```python
def compute_result(num: int) -> int:
    if num % 2 == 0:
        return num
    
    else:
        return num * 2
```

**Исправление**:

```python
def compute_result(num: int) -> int:
    if num % 2 == 0:
        return num

    return num * 2
```

## Ненужное ветвление в функциях

**Пример**:

```python
def is_even(num: int) -> bool:
    if num % 2 == 0:
        return True

    return False
```

**Исправление**:

```python
def is_even(num: int) -> bool:
    return num % 2 == 0
```

## Неправильная проверка коллекций на пустоту

**Пример**:

```python
nums_unique = set(range(5))

if len(nums_unique) != 0:
    print("collection is not empty")
```

**Исправление**:

```python
nums_unique = set(range(5))

if nums_unique:
    print("collection is not empty")
```
